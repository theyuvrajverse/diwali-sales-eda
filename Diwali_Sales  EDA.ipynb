# Gender

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
sns.countplot(x='Gender', data=df)
plt.show()
```

```python
ax = sns.countplot(x= 'Gender' ,data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

# Age Group

```python
sns.countplot(data=df,x='Age Group')
```

```python
ax = sns.countplot(x= 'Age Group' ,data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
bx= sns.countplot(data=df, x='Age Group', hue='Gender')

for bars in bx.containers:
    bx.bar_label(bars)
```

```python
df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values (by='Amount',ascending=False)
```

```python
# Total Amount vs Age Group
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values (by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

From Above Graphs we can see thet most of the buyers are of age group between 26 to 35 year Female

#  State

```python
# Total no. of order from top 10 states
```

```python
sale_state= df.groupby(['State'],as_index =False)['Orders'].sum().sort_values (by="Orders",ascending=False)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Orders',data=sale_state)
```

from the above graph we can see that most of the orders are booked by UP

```python
# Total No. of Amount from top 10 States.
```

```python
Amount_State=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Amount',data=Amount_State)
```

From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

# Marital Status

```python
cx = sns.countplot(data= df,x='Marital_Status')

sns.set(rc={'figure.figsize':(2,5)})
for bars in cx.containers:
    cx.bar_label(bars)
```

```python
Marital_Amount=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x='Marital_Status',y='Amount',hue='Gender',data=Marital_Amount)
```

From above graphs we can see that most of the buyers are married (women) and they have high purchasing power

# Product Category

```python
df.columns
```

```python
ex=sns.countplot(x='Product_Category',data=df)
sns.set(rc={"figure.figsize":(30,10)})

for bars in ex.containers:
    ex.bar_label(bars)
```

```python
product_cost=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by=["Amount"],ascending=False)
sns.set(rc={'figure.figsize':(30,10)})
sns.barplot(x='Product_Category',y='Amount',data=product_cost)
```

Looking at the graph it is clear that clothing & apparel are the most commonly ordered items
and the data also shows that Food has the highest purchasing power among customers

```python
most_selling_productID=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by=['Orders'],ascending=False)
sns.barplot(x='Product_ID',y='Orders',data=most_selling_productID)
```

```python
most_saling_productID=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by=['Orders'],ascending=False)
sns.barplot(x='Product_ID',y='Orders',data=most_saling_productID)
```

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```
